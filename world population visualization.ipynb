# Draw the  Graph of Top 5 countries based on the 2022 Population

```python
for x in data["Continent"].unique():
    plt.figure(figsize = (20,10))
    temp =  data[data["Continent"]==x][["Country/Territory", "2022 Population"]].sort_values(by = "2022 Population", ascending = False, ignore_index = True)[:5]
    plt.title(f"Top 5 countries based on the 2022 Population in {x}")
    sns.barplot(data = temp, x  = "Country/Territory", y = "2022 Population", palette = "Set1")
    plt.show()
```

```python
diff=  data.groupby("Continent")[["1970 Population", "2022 Population"]].sum().sort_values(by ="Continent" , ascending = True).reset_index()
```

```python
diff
```

#  Population comparision between 1970 and 2022 all over the world

```python
plt.figure(figsize = (16,8))
plt.plot(diff["Continent"], diff["1970 Population"], label = "1970", color = "red", marker = "o")
plt.plot(diff["Continent"], diff["2022 Population"], label = "2022", color = "green", marker = "o")
plt.xlabel("Continent")
plt.ylabel("Population")
plt.title("Population comparision between 1970 and 2022")
plt.grid()
plt.show()
```

# Analizing the trends of the population across the world between 1970 - 2022 

```python
plt.figure(figsize = (12,9))
population_feature = ['2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population']
population_feature.reverse()
for f in population_feature:
    plt.plot(conti_data[f], label = f)
plt.legend()
plt.grid()
plt.show()
```
